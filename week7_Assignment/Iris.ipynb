```python
# import libraries.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# TASK 1. Load and explore datasets.
try:
    iris_data = load_iris(as_frame=True)
    df = iris_data.frame
    print("Dataset loading successfully!")
except Exception as e:
    print(f"Error loading dataset: {e}")

# Display first rows.
print(df.head())

# Check data types and missing values.
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())

# There's no presence of any missing value in the iris dataset.

# TASK 2. Basic Data Analysis.
print("\nStatistical Summary:\n", df.describe())

# Group according to species and get their mean for each group.
grouped = df.groupby('target').mean()
print("\nMean values grouped by species:\n", grouped)

# TASK 3, Data Visualization.

# Map numeric target to species name.
df['species'] = df['target'].map(dict(enumerate(iris_data.target_names)))

# Line chart: mean measurements per species
grouped.plot(kind='line', title='Mean Measurements by Species')
plt.xlabel('Species')
plt.ylabel('Measurement')
plt.legend(loc='best')
plt.grid(True)
plt.show()

# Bar chart: average petal length per species
sns.barplot(x='species', y='petal length (cm)', data=df)
plt.title('Average Petal Length per Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()

# Histogram: sepal length distribution
plt.hist(df['sepal length (cm)'], bins=15, edgecolor='black')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()

# Scatter plot: Sepal Length vs Petal Length
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', hue='species', data=df)
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend()
plt.show()
```
